leap year

```
year = int(input("enter: "))
if (year % 400 == 0) & ( year % 100 == 0):
   print("{}leap year".format(year))
elif ( year % 4 == 0) & ( year % 100 != 0):
    print("{} leap year".format(year))
else:
    print("{}not a leap year".format(year))
```

largest among three numbers


```
first_num= float(input("enter a first number: "))
second_num= float(input("enter a second number: "))
third_num= float(input("enter a third number: "))
if(first_num > second_num ) and (first_num > third_num):
  print("{} is greater".format(first_num))
elif(second_num > third_num):
  print("{} is greater".format(second_num))
else:
  print("{} is greater".format(third_num))
```

prime series

```
start = int(input("enter a start: "))
end = int(input("enter a end value: "))
for i in range(start,end):
  for j in range (2,i):
    if i % j == 0:
      break
  else:
      print(i,end=" 1")
```

check prime

```
s = int(input("enter a num: "))
for i in range (2,s):
  if s % i == 0:
    print(s, "is not a prime")
    break
else:
    print(s, "is a prime")
```

factorial

```
num = int(input("enter the value: "))
fact = 1
if num < 0:
  print("factorial does not exist")
elif(num == 0):
  print("factorial of 0 is 1")
else:
  for i in range ( 1,num+1 ):
    fact = fact*i
  print("the factorial of ",num,"is " ,fact)
```
